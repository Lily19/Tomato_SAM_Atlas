```r
mycolor <- c(
  '#E41A1C', '#377EB8', '#4DAF4A', '#984EA3', '#FF7F00', 
  '#FFFF33', '#A65628', '#F781BF', '#66C2A5', '#FC8D62', 
  '#8DA0CB', '#E78AC3', '#A6D854', '#FFD92F', '#E5C494', 
  '#B3B3B3', '#1B9E77', '#D95F02', '#7570B3', '#E7298A', 
  '#66A61E', '#E6AB02', '#A6761D', '#666666', '#A6CEE3', 
  '#1F78B4', '#B2DF8A', '#33A022'
)
```

```r
options(repr.plot.width=6, repr.plot.height=5)
pic1 <- DimPlot(
    Data, 
    group.by = 'seurat_clusters', 
    label=TRUE, 
    label.size = 6, 
    reduction='umap',
    cols=mycolor
) + ggtitle("")
pic1
```

```r
options(repr.plot.width=6, repr.plot.height=5)
pic2 <- DimPlot(
    Data, 
    group.by = 'assign.ident', 
    label=TRUE, 
    label.size = 6, 
    reduction='umap',
    cols=mycolor
) + ggtitle("")
pic2
```

```r
pic1 <- DimPlot(
    Data, 
    group.by = 'seurat_clusters', 
    label = TRUE,
    label.size = 6,                # 增大标签字号
    label.box = TRUE,              # 添加标签背景框
    cols = mycolor,                # 使用自定义颜色
    pt.size = 1.2,                 # 调整点大小（默认1.0）
    # pt.alpha = 0.8,                # 设置透明度（0-1）
    order = "shuffle",             # 打散点绘制顺序避免覆盖
    repel = TRUE,                  # 防止标签重叠
    raster = TRUE                  # 启用光栅化提升渲染速度
) + 
  ggtitle("") +                    # 移除默认标题
  theme_minimal(base_size = 14) +  # 设置基础字体大小
  theme(
    aspect.ratio = 1,              # 强制1:1比例
    legend.position = "none",      # 移除图例（标签已显示）
    axis.title = element_blank(),  # 移除坐标轴标题
    axis.text = element_blank(),   # 移除坐标轴刻度
    axis.ticks = element_blank(),  # 移除坐标轴刻度线
    panel.grid = element_blank()   # 移除背景网格线
  )
```

```r
pic2 <- DimPlot(
    Data, 
    group.by = 'assign.ident', 
    label = TRUE,
    label.size = 6,                # 增大标签字号
    label.box = TRUE,              # 添加标签背景框
    cols = mycolor,                # 使用自定义颜色
    pt.size = 1.2,                 # 调整点大小（默认1.0）
    # pt.alpha = 0.8,                # 设置透明度（0-1）
    order = "shuffle",             # 打散点绘制顺序避免覆盖
    repel = TRUE,                  # 防止标签重叠
    raster = TRUE                  # 启用光栅化提升渲染速度
) + 
  ggtitle("") +                    # 移除默认标题
  theme_minimal(base_size = 14) +  # 设置基础字体大小
  theme(
    aspect.ratio = 1,              # 强制1:1比例
    legend.position = "none",      # 移除图例（标签已显示）
    axis.title = element_blank(),  # 移除坐标轴标题
    axis.text = element_blank(),   # 移除坐标轴刻度
    axis.ticks = element_blank(),  # 移除坐标轴刻度线
    panel.grid = element_blank()   # 移除背景网格线
  )
```
