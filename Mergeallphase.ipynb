```r
mycolor <- c(
  '#E41A1C', '#377EB8', '#4DAF4A', '#984EA3', '#FF7F00', 
  '#FFFF33', '#A65628', '#F781BF', '#66C2A5', '#FC8D62', 
  '#8DA0CB', '#E78AC3', '#A6D854', '#FFD92F', '#E5C494', 
  '#B3B3B3', '#1B9E77', '#D95F02', '#7570B3', '#E7298A', 
  '#66A61E', '#E6AB02', '#A6761D', '#666666', '#A6CEE3', 
  '#1F78B4', '#B2DF8A', '#33A022'
)
```

```r
options(repr.plot.width=6, repr.plot.height=5)
pic1 <- DimPlot(
    Data, 
    group.by = 'seurat_clusters', 
    label=TRUE, 
    label.size = 6, 
    reduction='umap',
    cols=mycolor
) + ggtitle("")
pic1
```

```r
options(repr.plot.width=6, repr.plot.height=5)
pic2 <- DimPlot(
    Data, 
    group.by = 'assign.ident', 
    label=TRUE, 
    label.size = 6, 
    reduction='umap',
    cols=mycolor
) + ggtitle("")
pic2
```

```r
pic1 <- DimPlot(
    Data, 
    group.by = 'seurat_clusters', 
    label = TRUE,
    label.size = 6,                # 
    label.box = TRUE,              # 
    cols = mycolor,                # 
    pt.size = 1.2,                 # 1.0
    # pt.alpha = 0.8,                # 0-1
    order = "shuffle",             # 
    repel = TRUE,                  # 
    raster = TRUE                  # 
) + 
  ggtitle("") +                    # 
  theme_minimal(base_size = 14) +  # 
  theme(
    aspect.ratio = 1,              # 1:1
    legend.position = "none",      # 
    axis.title = element_blank(),  # 
    axis.text = element_blank(),   # 
    axis.ticks = element_blank(),  # 
    panel.grid = element_blank()   # 
  )
```

```r
pic2 <- DimPlot(
    Data, 
    group.by = 'assign.ident', 
    label = TRUE,
    label.size = 6,                # 
    label.box = TRUE,              # 
    cols = mycolor,                # 
    pt.size = 1.2,                 # 1.0
    # pt.alpha = 0.8,                # 0-1
    order = "shuffle",             # 
    repel = TRUE,                  # 
    raster = TRUE                  # 
) + 
  ggtitle("") +                    # 
  theme_minimal(base_size = 14) +  # 
  theme(
    aspect.ratio = 1,              # 1:1
    legend.position = "none",      # 
    axis.title = element_blank(),  # 
    axis.text = element_blank(),   # 
    axis.ticks = element_blank(),  # 
    panel.grid = element_blank()   # 
  )
```
